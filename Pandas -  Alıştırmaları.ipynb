```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1 - Seaborn içinde titanic datasetini tanımla
df = sns.load_dataset("titanic")
df.head()
```

```python
# 2 - Titanic verisetindeki kadın ve erkek yolcu sayısını bulunuz
sex_count = df["sex"].value_counts()
print(sex_count)
```

```python
# 3 - Her bir sütuna ait unique değerlerin sayısını bulunuz.
unique_count = df.nunique()
print(unique_count)
```

```python
# 4 - Plcass değişkeninin unique değerlerinin sayısını bulunuz
unique_count_plcass = df["pclass"].nunique()
print(unique_count_plcass)
```

```python
# 5 - pclass ve parch değişkenlerinin unique değerlerinin sayısını bulunuz
unique_count_plcass_parch = df[["pclass", "parch"]].nunique()
print(unique_count_plcass_parch)
```

```python
#6 - Embarked Değişkeninin tipini kontrol ediniz. Sonrasında tipini category olarak değiştirip tekrar kontrol ediniz.
print(df["embarked"].dtype)
df["embarked"] = df["embarked"].astype("category")

print(df["embarked"].dtype)
```

```python
# 7 - Embarked Değeri C olanların tüm bilgilerini gösteriniz

df[df["embarked"] == "C"]
```

```python
# 8 - Embarked Değeri S olmayanların tüm bilgilerini gösteriniz

df[df["embarked"] != "S"]
```

```python
# 9 - Yaşı 30'dan küçük ve kadın olan yolcuların tüm bilgilerini gösteriniz.
df.loc[(df["age"] < 30) & (df["sex"] == "female"), :]
```

```python
# 10 - Fare'i 500 den büyük veya yaşı 70'den büyük yolcuların bilgilerini gösteriniz.
df.loc[(df["fare"] > 500) | df["age"] > 70.0, :]
```

```python
# 11 - Her bir değişkendeki boş değerlerin toplamını bulunuz.
df.isnull().sum()
```

```python
# 12 - Who değişkenini dataframe'den çıkartınız
df.drop("who", axis = 1, inplace = True)
df
```

```python
# 13 - Deck değişkenindeki boş değerleri deck değişkeninin en çok tekrar eden değeri (mode) ile doldurunuz.
df["deck"].isnull().sum()

df["deck"] = df["deck"].fillna(df["deck"].mode()[0])
```

```python
df["deck"].isnull().sum()
```

```python
# 14 - Age değişkenindeki boş değerleri age değişkeninin medyanı ile doldurunuz.
df["age"] = df["age"].fillna(df["age"].median())
```

```python
# 15 - Survived değişkeninin pclass ve cinsiyet değişkenleri kırılımında sum, count ve mean değerlerini bulunuz.
df.groupby(["pclass", "sex"]).agg({
    "survived": ["sum", "count", "mean"]
})
```

```python
# 16 - 30 yaşında altında olan 1; 30'a eşit ve üstünde olanlara 0 verecek bir fonk. yaz.
# Yazılan fonk. kullanarak titanik veri setinde age_flag adında bir değişken oluştur. (apply & lambda kullanılacak)

def age_check(row):
    if row < 30:
        return 1
    else:
        return 0

df["age_flag"] = df["age"].apply(lambda x: age_check(x))
df.head()
```

```python
# 2. yol -U
df["age_flag"] = df["age"].apply(age_check)
df.head()
```

```python
#17 - Seaborn kütüphanesi içerisinden tips datasetini tanımla
df = sns.load_dataset("tips")
df.head()
```

```python
# 18 - Time değişkeninin kategorilerine göre total_bill değerinin sum, min, max ve mean değerlerini bul.

df.groupby(["time"]).agg({
    "total_bill": ["sum", "min", "max", "mean"]
})
```

```python
# 19 - Day ve time'a göre total_bill değerlerinin sum, min, max ve mean değerlerini bul.

df.groupby(["day", "time"]).agg({
    "total_bill": ["sum", "min", "max", "mean"]
})
```

```python
#20 - Lunch zamanına ve kadın müşterilere ait total_bill değerlerinin day'e göre sum, min, max ve mean değerlerini bul.

# Lunch zamanı ve kadın müşteriler
df_filtered = df.loc[(df["time"] == "Lunch") & (df["sex"] == "Female"), :]

df_filtered.agg({
    "total_bill": ["sum", "min", "max", "mean"]
})
```

```python
#21- size'ı 3'den küçük, total_bill'i 10 dan büyük olan siparişlerin ortalaması (loc)

df_filtered = df.loc[(df["size"] < 3) & (df["total_bill"] > 10), :]

df_filtered.total_bill.mean()
```

```python
# 22 - Total_bill_tip_sum adında yeni bir değişken oluştur. Her bir müşterinin ödediği total bill ve tip'in toplamını versin

df["total_bill_tip_sum"] = df["total_bill"] + df["tip"]
df.head()
```

```python
# 23 - Total_bill değişkeninin kadın ve erkek için ayrı ayrı ortalamasını bulunuz.
# Bulduunuz ortalamarın altında olanlar 0, üstünde ve eşit olanlara 1 verildiği yeni bir total_bill_flag değişkeni oluşturunuz.

# Kadınlar için female ortalamaları, erkekler için de male ortalamaları dikkate alınacaktır. 
# Param olarak cinsiyet ve total_bill alan bir fonksiyon yazarak başlayınız (if - else içeriecek.)

def mean_by_sex(sex, total_bill):
    mean = df.loc[df["sex"] == sex, ["total_bill"]].mean()[0]
    if total_bill < mean:
        return 0
    else:
        return 1

    
df["total_bill_flag"] = df.apply(lambda x: mean_by_sex(x["sex"], x["total_bill"]), axis = 1)

df.head()
# df["total_bill_flag"].value_counts()
```

```python
#23 - 80 kat Daha hızlı hali

mean_male = df.loc[df["sex"] == "Male", ["total_bill"]].mean()[0]
mean_female = df.loc[df["sex"] == "Female", ["total_bill"]].mean()[0]

def mean_by_sex(sex, total_bill):        
    mean = mean_male if sex == "Male" else mean_female
        
    if total_bill < mean:
        return 0
    
    else:
        return 1

    
df["total_bill_flag"] = df.apply(lambda x: mean_by_sex(x["sex"], x["total_bill"]), axis = 1)

df.head()
# df["total_bill_flag"].value_counts()
```

```python
# 24 - total_bill_flag değişkeninini kullanarak cinsiyete göre ortalamanın altında ve üstüne olanların sayısını gözlemleyiniz

df.groupby(["sex", "total_bill_flag"]).agg({
    "total_bill_flag" : ["count"]
})
```

```python
# 25 - veriyi total_bill_tip_sum değişkenine göre büyükten küçüğe sıralayınız ve ilk 30 kişiyi yeni bir dataframe'e atayınız.

new_df = df.sort_values("total_bill_tip_sum", ascending = False)[0:30]
```
