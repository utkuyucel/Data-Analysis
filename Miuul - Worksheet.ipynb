```python
# Mülakat sorusu
# Amaç: Aşağıdaki şekilde tring değiştiren fonksiyon yazmak istiyoruz.
#
# input: "hi my name is john and i am learning python"
# outuput: "Hi mY NaMe iS JoHn aNd i aM LeArNiNg pYtHoN"
#
# Verilen string ifadelerdeki çift indextekileri büyüt, tek indekstekileri küçült.
```

```python
input_string = "hi my name is john and i am smashing python"
len(input_string)
```

```python
def alternating(string):
    out = ""
    
    for index in range(0, len(string)):
        if index % 2 == 0:
            out += input_string[index].upper()        
        else:
            out += input_string[index].lower()            
    print(out)

alternating(input_string)
```

```python
def alternating_2(string):
    out = ""
    
    for index, val in enumerate(string):
        if index % 2 == 0:
            out += val.upper()
        
        else:
            out += val.lower()
    
    print(out)
        
alternating_2(input_string)
```

```python
# divide_students fonksiyonu yazılacak.
# Çift indexte yer alan öğrencileri bir listeye
# Tek indexte yer alan öğrencileri başka bir listeye alınız.
# Fakat bu iki liste tek bir liste olarak return olsun.

students = ["John", "Mark", "Venessa", "Mariam"]
```

```python
def divide_students(input_list: list) -> list:
    groups = [[], []]
    for idx, name in enumerate(input_list):
        if idx % 2 == 0:
            groups[0].append(name)
        else:
            groups[1].append(name)
            
    return groups
    
output = divide_students(students)
print(output)
```

```python
#Categoric, Numeric ve Categoric but cardinal listeleri veren bir fonksiyon yazmamız gerekiyor.
# cat_th, bir değişken int olsa bile eğer nunique < cat_th; bu kategorik değişkendir.
# car_th, bir değişken categorical olsa bile eğer nunique > car_th; bu car değişkendir.
# TOPARLAMA

def grab_col_names(dataframe, cat_th = 10, car_th = 20):
    """
    Gives us the numeric, categorical and categorical but cardinal variables from given dataframe
    
    Parameters
    ----------
    dataframe: dataframe
        The dataframe which we want to get variable names from.
        
    cat_th: int, float
        Threshold value for numerical but categorical variables
        
    car_th: int, float
        Threshold value for categorical but cardinal variables.
        
    Returns
    -------
    cat_cols: list
        List of categorical variables
            
    num_cols: list
        List of numeric variable

    cat_but_car: list
        List of categorical but cardinal variables
    
    Notes
    -----
    cat_cols + num_cols + cat_but_car = total number of variables
    num_but_cat is in cat_cols.
    
    """
    # Categorical
    cat_cols = [col for col in df.columns if str(df[col].dtypes) in ["category", "object", "bool"]]
    num_but_cat = [col for col in df.columns if df[col].nunique() < 10 and df[col].dtypes in ["int64", "float64", "int32", "float32"]]
    
    cat_but_car = [col for col in df.columns if df[col].nunique() > 20 and df[col].dtypes in ["category", "object"]]

    cat_cols = cat_cols + num_but_cat # Cat cols güncellendi
    cat_cols = [col for col in cat_cols if col not in cat_but_car]
    
    
    # Numeric
    num_cols = [col for col in df.columns if df[col].dtypes in ["int64", "float64", "int32", "float32"]] #numeric olanları al
    num_cols = [col for col in num_cols if col not in cat_cols] # ne olur ne olmaz, categorical olandan çıkart.
    
    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables: {dataframe.shape[1]}")
    print(f"cat_cols: {len(cat_cols)}")
    print(f"num_cols: {len(num_cols)}")
    print(f"cat_but_car: {len(cat_but_car)}")
    print(f"num_but_car: {len(num_but_cat)}")
    
    return cat_cols, num_cols, cat_but_car
```

```python
## Korelasyon Analizi (Correlation Analysis)

df = pd.read_csv("archive/data.csv")
df = df.iloc[:, 1:-1] # id ve en sondaki nan value olan işe yaramaz kısmı atıyoruz

df.head()
```

```python
# Yüksek korelasyonu olan değişkenlerin silinmesi.
corr_matrix = df.corr().abs() # Tüm korelasyonlar pozitif hale getiriliyor. / Negatif ve pozitif yüksek kor. aynı anlamdadır ve ikisinin de gitmesi gerekiyor.

#Çoklama / Birbirini tekrar eden değişkenlerden kurtulduk.
upper_triangle_matrix = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k = 1).astype(bool))
upper_triangle_matrix
```

```python
drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > 0.90)]

drop_list # 90%'dan yüksek korelasyona sahip değişkenler.
```

```python
corr_matrix[drop_list] #Yüksek korelasyonlu olanlar (en az 1 tanesi yüksek)
```

```python
df.drop(drop_list, axis = 1) # Silindi
```

```python
def high_correlated_cols(dataframe, plot = False, corr_th = 0.90):
    corr = dataframe.corr() #korelasyon oluşturduk
    corr_matrix = corr.abs() #Mutlak Değerini aldık
    upper_triangle_matrix = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k = 1).astype(bool))
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]
    
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc = {"figure.figsize": (15,15)})
        sns.heatmap(corr, cmap = "RdBu")
        plt.show()
    
    return drop_list
```

```python
drop_list = high_correlated_cols(df, plot = True)
df.drop(drop_list, axis = 1)
```

```python
# Öğrendik, veri geldiğinde uygulayalım değil ancak bunu öğrenelim. 
# Koyu ve yoğun kısımlar gitti.
high_correlated_cols(df.drop(drop_list, axis = 1), plot = True)
```
