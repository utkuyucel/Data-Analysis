# Görev 2 - Country, source, sex, age kırılımında ortalama kazançlar nedir?

```python
df.groupby(["COUNTRY", "SOURCE", "SEX", "AGE"]).agg({
    "PRICE" : "mean"
})
```

# Görev 3 - Çıktıyı PRICE'a göre sıralayınız

```python
# önceki sorudakki çıktıyı daha iyi görebilmek için sort_values metodunu azalan olacak şekilde PRICE'a göre uygulayın
# çıktıyı agg_df olarak kaydedin.

agg_df = df.groupby(["COUNTRY", "SOURCE", "SEX", "AGE"]).agg({
    "PRICE" : "mean"
}).sort_values("PRICE", ascending = False)

agg_df.head()
```

# Görev 4 - Indexte yer alan isimleri değişken ismine çeviriniz.

```python
# 3. sorunun çıktısında yer alan PRICE dışındaki tüm değişkenler index isimleridir. bu isimleri değişken ismine çeviriniz.

agg_df = agg_df.reset_index()
agg_df.head()
```

# 5 - Age değişkenini kategorik değişkene çeviriniz ve agg_df'e ekleyiniz.

```python
# Age sayısal değişkenini kategorik değişkene çevir
# Aralıkları ikna edici şekilde oluştur. (0_18,19_23, 24_30, 31_40, 41_70)

agg_df["AGE_CAT"] = pd.cut(
    x = agg_df["AGE"],
    bins = [0, 18, 23, 30, 40, 70],
    labels = ["0_18", "19_23", "24_30", "31_40", "41_70"]
)


# agg_df["AGE_CAT"].value_counts()
agg_df.head()
```

# 6- Yeni seviye tabanlı müşterileri (persona) tanımlayınız.

```python
# Yeni seviye tabanlı müşterileri (Persona) tanımlayınız ve veri setine değişken olarak ekleyiniz.
# Yeni eklenecek değişken adı: customers_level_based

# Önceki soruda elde edeceğiniz çıktıdaki gözlemleri bir araya getirerek customers_level_based
# değişkenini oluşturmanız gerekmektedir.

# Dikkat: List comprehension ile customers_level_based değerleri oluşturduktan sonra bu değerlerin tekilleştirilmesi gerekiyor.
# USA_ANDROID_MALE_0_18 birden fazla varsa bunları groupby'a alıp price ortalamalarını almak lazım.

agg_df = pd.DataFrame([{
    "customers_level_based" : (str(i[0]) + "_" + str(i[1]) + "_" + str(i[2]) + "_" + str(i[5])).upper(), "PRICE": i[4]}
    for i in agg_df.values
])
```

```python
agg_df = agg_df.groupby("customers_level_based").agg({
    "PRICE":"mean"
})

agg_df
```

# Görev 7 - Yeni Müşterileri (Personaları) segmentlere ayırınız.

```python
# Yeni müşterileri (USA_ANDROID_MALE_0_18 mesela), price'a göre 4 segmente ayırınız.

# Segmentleri SEGMENT isimlendirmesi ile değişkken olarak agg_df'e ekleyiniz.

# Segmentleri betimleyiniz (Segmentlere göre group by yapıp price mean max ve sum'larını alınız.)
```

```python
agg_df["SEGMENT"] = pd.cut(agg_df["PRICE"], 4, labels = ["D", "C", "B","A"])
```

```python
agg_df = agg_df.reset_index()
agg_df.head()
# agg_df.value_counts()
```

```python
agg_df.groupby("SEGMENT").agg({
    "PRICE": ["mean", "max", "sum"]
})
```
